## Количество обзоров и средняя оценка.

Для каждой книги посчитать количество обзоров и среднюю оценку.

```python
query = '''SELECT
               books.book_id AS book_id,
               books.title AS title,
               AVG(ratings.rating) AS mean_rating,
               COUNT(reviews.review_id) AS review_count
           FROM
               books
               INNER JOIN ratings ON ratings.book_id = books.book_id
               INNER JOIN reviews ON reviews.book_id = books.book_id
           GROUP BY
               books.book_id,
               books.title
           ORDER BY
               mean_rating DESC,
               review_count DESC;    
        '''
books_rating = pd.io.sql.read_sql(query, con=engine)
pd.io.sql.read_sql(query, con = engine)
```

Средняя оценка книг составляет от 5 до 1.5 баллов.

## Издательство, которое выпустило наибольшее число книг толще 50 страниц.

Определить издательство, которое выпустило наибольшее число книг толще 50 страниц.

```python
query = '''SELECT
               publishers.publisher_id AS publisher_id,
               publishers.publisher AS publisher,
               COUNT(DISTINCT books.book_id) AS books_count
           FROM
               publishers
               INNER JOIN books ON publishers.publisher_id = books.publisher_id
           WHERE
               books.num_pages > 50
           GROUP BY
               publishers.publisher_id,
               publishers.publisher
           ORDER BY
               books_count DESC
           LIMIT 1;
        '''
publishers_books = pd.io.sql.read_sql(query, con=engine)
pd.io.sql.read_sql(query, con = engine)
```

Издательство Penguin Books выпустило максимальное количество книг толще 50 страниц из базы сервиса - 42.

## Автор, с самой высокой средней оценкой книг.

Определить автора с самой высокой средней оценкой книг — учитывать только книги с 50 и более оценками.

Исходя из формулировки запроса считаю, что в данном случае нужно определить среднюю оценку каждой книги, а потом определить среднюю оценку книг по каждому автору.

```python
query = '''SELECT
               authors.author_id AS author_id,
               authors.author AS author,
               AVG(Sub.rating_mean) AS mean_rating
           FROM (
               SELECT
                   books.book_id AS book_id,
                   books.author_id AS author_id,
                   COUNT(DISTINCT ratings.rating_id) AS ratings_count,
                   AVG(ratings.rating) AS rating_mean
               FROM
                   ratings
                   INNER JOIN books ON ratings.book_id = books.book_id
               GROUP BY
                   books.book_id,
                   books.author_id
               HAVING
                   COUNT(DISTINCT ratings.rating_id) > 49) AS Sub
               LEFT JOIN authors ON Sub.author_id = authors.author_id
           GROUP BY
               authors.author_id,
               authors.author
           ORDER BY
               mean_rating DESC
           LIMIT 1;
        '''
big_author = pd.io.sql.read_sql(query, con=engine)
pd.io.sql.read_sql(query, con = engine)
```

Автор книг с самой высокой средней оценкой - Дж.К. Роулинг, иллюстратор этих книг - Мэри ГранПре, скорее всего это американские из дания книг о Гарри Поттере.

## Обзоры пользователей.

Посчитать среднее количество обзоров от пользователей, которые поставили больше 50 оценок.

```python
query = '''SELECT
               username,
               AVG(count_reviews) AS mean_cnt_review
           FROM (
               SELECT
                   SubQ.username AS username,
                   COUNT(reviews.review_id) AS count_reviews
               FROM (
                   SELECT
                       username,
                       COUNT(rating_id) AS rating_count
                   FROM
                       ratings
                   GROUP BY
                       username
                   HAVING
                       COUNT(rating_id) > 50) AS SubQ
               LEFT JOIN reviews ON reviews.username = SubQ.username
               GROUP BY
                   SubQ.username) AS SubQ1
           GROUP BY
               username
           ORDER BY
               mean_cnt_review DESC;
        '''
users_reviews = pd.io.sql.read_sql(query, con=engine)
pd.io.sql.read_sql(query, con = engine)
```

Пользователей, которые поставили больше 50 оценок всего 6, при этом эти же активные пользователи пишут большое количество отзывов на книги.
